```python
import pandas as pd
import plotly.express as px
```

```python
top_ten_marketcap =pd.read_csv("10_US_comp_marketcap.csv")
print(top_ten_marketcap)
```

```python
fig = px.scatter(top_ten_marketcap, "Date", "Volume NVDA")
fig.show()
```

```python
fig = px.scatter(top_ten_marketcap, "Close TSLA", "Volume TSLA", color="Volume TSLA",
                 hover_data=["Volume TSLA"])
fig.show()
```

```python
fig = px.line(top_ten_marketcap, x='Date', y="Open AAPL")
fig.show()
fig = px.line(top_ten_marketcap, x='Date', y="Close AAPL")
fig.show()
```

```python
px.line(top_ten_marketcap, x='Date', y=["Open TSLA", "Close TSLA"], 
        labels={"x":"Date", "y": "Stock price"},
       title="Opening and closing price for Tesla stock 2022")
```

```python
px.line(top_ten_marketcap, x='Date', y=["Close AAPL", "Close MSFT", "Close GOOG",
                                        "Close AMZN", "Close NVDA", "Close BRKB", 
                                        "Close TSLA", "Close META", "Close JNJ", "Close V"], 
        labels={"x":"Date", "y":"Stock price"},
        title="Top 10 US companies by marketcap 2022")
```

```python
px.line(top_ten_marketcap, x='Date', y=["Close AAPL", "Close MSFT", "Close GOOG",
                                        "Close AMZN", "Close NVDA", "Close BRKB", 
                                        "Close TSLA", "Close META", "Close JNJ", "Close V"], 
        labels={"x":"Date", "y":"Stock price"},
        title="Top 10 US companies by marketcap 2022")
```

```python

fig = px.scatter(top_ten_marketcap, x='Date', y=["Volume AAPL", "Volume MSFT", "Volume GOOG",
                                        "Volume AMZN", "Volume NVDA", "Volume BRKB", 
                                        "Volume TSLA", "Volume META", "Volume JNJ", "Volume V"],
        color="Daily Volume", hover_data=["Date"])
fig.show()
```

```python
top_ten_monthly_data =pd.read_csv("10_US_comp_marketcap_monthly.csv")
print(top_ten_monthly_data)
```

```python
fig = px.scatter(top_ten_monthly_data, x='Date', y=["Volume AAPL", "Volume MSFT", "Volume GOOG",
                                        "Volume AMZN", "Volume NVDA", "Volume BRKB", 
                                        "Volume TSLA", "Volume META", "Volume JNJ", "Volume V"],
        color="Daily Volume", hover_data=["Date"])
fig.show()
```

```python
fig = px.line(top_ten_marketcap, x='Date', y="Daily Volume", markers=True)
fig.show()
```

```python
fig = px.bar(top_ten_marketcap, x='Date', y="Daily Volume")
fig.show()
```

```python
fig = px.bar(top_ten_monthly_data, x='Date', y="Daily Volume")
fig.show()
```

```python
fig = px.box(top_ten_marketcap, y=["Volume AAPL", "Volume AMZN", 
                                   "Volume NVDA", "Volume TSLA", "Volume META"])
fig.show()
```

```python
fig = px.box(top_ten_monthly_data, y=["Volume AAPL", "Volume AMZN", 
                                   "Volume NVDA", "Volume TSLA", "Volume META"], points="all")
fig.show()
```

```python
fig = px.bar(top_ten_monthly_data, x='Date', y="Close TSLA")
fig.show()
```

```python
fig = px.bar(top_ten_monthly_data, x='Date', y="Close TSLA", color="Close TSLA")
fig.show()
```

```python
fig = px.bar(top_ten_monthly_data, x='Date', y="Close TSLA", color="Close TSLA", 
            text_auto="0.5s", pattern_shape_sequence="x")
fig.update_traces(textfont_size=13, textangle=270, textposition="outside")
fig.show()
```
